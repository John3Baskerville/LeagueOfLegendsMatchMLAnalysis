# Objective
League of Legends is a competitive online battle arena where two teams of 5 players and their unique characters face off on a map. The objective of each match is to infiltrate the enemy team base and destroy its nexus. Each match lasts around 30 minutes and the depth of gameplay requires cooperation, management of resources, and strategy in order to win a match. 
The objective of this project is to use a ranked game dataset and its features to categorize the results of a match as either a win or a loss for either the red or blue team. 

#### Source URL
Dataset URL: https://www.kaggle.com/bobbyscience/league-of-legends-diamond-ranked-games-10-min

```python
#import pandas to handle dataset in easy to use table format
import pandas as pd
#preprocessing package from sklearn will normalize data, making it easier for keras and model to learn
from sklearn import preprocessing
#train_test_split function for easy validation splits
#GridSearch will be used to select optimal parameters
from sklearn.model_selection import train_test_split, GridSearchCV
#import keras libraries to build model and manipulate layers
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD
#import plotting libraries
from matplotlib import pyplot as plt
#use datetime to track time spent training
import datetime
#import numpy for data preprocessing and random seed setting
import numpy as np
#if needed, will transform y variable.
from keras.utils import to_categorical
#seaborn library to help visualize correlations
import seaborn as sns
```

# Input variables and Preprocessing steps
There 2 teams with 19 features each. Each row represents 1 game with collected data from the first 10 minutes. I expected some of the features to be highly correlated. Events such as takedowns of other players and completing in-game objectives all grant gold and gold is one of the most important resources to help characters progress in each match. 

After looking at a heatmap that showed correlations between features it was easy to remove features that were complicating the model. To shrink the number of features the final model used to train, I removed one of each couple features that had a correlation score higher than 0.8 or smaller than -0.8.

The dataset has 9,879 rows all used to train the models.

The column ‘blueWins’ is the dependent variable and a value of 1 represents a win for the blue team. 

```python
#store raw data in 'data' variable
data = pd.read_csv("high_diamond_ranked_10min.csv")

#create scaler function that will compress data to reduce varaibilty between features
min_max_scaler = preprocessing.MinMaxScaler()

#isolate dependant variable
y = data['blueWins'] # binary data
#Y = to_categorical(Y)
#First model will use all features except dependant variable and gameID
X = data.drop(['blueWins','gameId'],axis=1)
#use min max function to transform data
X = min_max_scaler.fit_transform(X)
#use train_test_split function to separate data for validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)

#dataShrink variable is new dataset with all highly correlated features removed
# dataShrink = data.drop(['blueWins','gameId','redFirstBlood','blueTotalGold','blueGoldPerMin','redDeaths',
#                   'redKills','redTotalGold','redGoldPerMin','blueCSPerMin',
#                   'blueExperienceDiff','redCSPerMin','redGoldDiff',
#                   'blueTotalExperience','redTotalExperience','redExperienceDiff',
#                    'blueAssists','redAssists','redEliteMonsters'],axis=1)
# data1 = min_max_scaler.fit_transform(dataShrink)
#X1_train, X1_test, Y1_train, Y1_test = train_test_split(data1, Y, test_size=0.18, random_state=42)
```

# Feature Selection
There are 38 features total, so the initial heatmap looks cluttered, but there are clear and obvious features that can be easily be removed. A feature that is highly correlated to another is likely to have the same effect on the dependent variable and likely is not needed. 

Now with a more simple model I can focus on improvements to the data and the neural network to increase accuracy.

```python
#create figure for heatmap plot
plt.figure(figsize = (15,10)) # width x height
#create correlation matrix with data
corr = data.corr()
#use heatmatp function to visualize correlations
sns.heatmap(corr,cmap='coolwarm',annot=False,linecolor='black',linewidth=1)
```

# ANN model

```python
seed = 100
np.random.seed(seed)

def build_model():
    model=Sequential()
    #input layer
    model.add(Dense(5,input_dim=38,activation='softmax'))
    model.add(Dense(250))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(14))
    model.add(Dense(35))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(1))
    model.add(Activation('sigmoid'))
    #optimizer = SGD(lr=.001, momentum=.5)
    #compile model with appropriate metrics
    model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['binary_accuracy','acc']) 
    return model

#the model will flatten before this limit
epochs = 250
#accurate and speedy batch size
batch_size = 100
#model parameterization used to help optimize parameters
model = KerasClassifier(build_fn=build_model)
#fit model
modelFit = model.fit(X_train,y_train,epochs=epochs, batch_size=batch_size,verbose=0,validation_data=(X_test,y_test))   

#plot loss and accuracy of trained model.
plt.figure()
plt.plot(modelFit.history['binary_accuracy'], color='blue', label='train accuracy')
plt.plot(modelFit.history['val_binary_accuracy'], label='test accuracy')
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend()
plt.title("Model Results: Accuracy")
```

# Results
When a confusion matrix is applied to the predicted results of our final model we can conclude an accuracy of about 72% when it comes to predicting a games outcome on our selected features. 3806 games were correctly predicted as wins and 3387 games were correctly predicted as losses. 

```python
from sklearn.metrics import confusion_matrix, accuracy_score

prediction = model.predict(X_test)
cm = confusion_matrix(y_test,prediction)
score = accuracy_score(y_test,prediction)
print("Confusion Matrix \n",cm)
print("\nModel Accuracy Score: ",score)
```
